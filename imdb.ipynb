#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

#### Option 3: calculate the average star rating for all genres, then filter using a boolean Series

```python
#First we take all the means, and create a variable for that dataframe that we can call
meanrating_genre = movies.groupby('genre')['star_rating'].mean()
#Second is we create a boolean series that reads True of the value counts of the genre are greater than 10
bool_series = movies['genre'].value_counts() > 10
#Next we select only those values from the genre means that are true in the boolean series.
meanrating_genre[bool_series]
```

#### Option 4: aggregate by count and mean, then filter using the count

```python
# Answer:
count_mean_genres = movies.groupby('genre')['star_rating'].agg(['count','mean'])
count_mean_genres
```

```python
##Now all we have to do is filter by the counts
relevant = count_mean_genres.loc[:,'count'] > 10
relevant_genres_means = count_mean_genres.loc[relevant,:]
relevant_genres_means
```
